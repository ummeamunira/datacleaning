```
%matplotlib notebook
import numpy as np
import pandas as pd
```

```
url1='https://www.eia.gov/outlooks/steo/xls/STEO_m.xlsx'
data=pd.read_excel(url1, sheet_name='2tab', header=[2,3], index_col=0).T.reset_index().drop(0)
m = {'Jan':'1', 'Feb':'2', 'Mar':'3', 'Apr':'4', 'May':'5', 'Jun':'6', 'Jul':'7', 'Aug':'8', 'Sep':'9', 'Oct':'10', 'Nov':'11', 'Dec':'12' }
data.level_1 = data.level_1.map(m)
data['yearmonth']=data['level_0'].map(str)+'m'+data['level_1'].map(str)
yearmonth=data['yearmonth']
WTIUS=data['WTIPUUS']
NGPUS=data['NGHHMCF']
op=pd.DataFrame({'yearmonth':yearmonth, 'WTIUS':WTIUS, 'NGPUS':NGPUS})
op
```

```
data1=pd.read_excel(url1, sheet_name='3atab', header=[2,3], index_col=0).T.reset_index().drop(0)
data1.level_1 = data1.level_1.map(m)
data1['yearmonth']=data1['level_0'].map(str)+'m'+data1['level_1'].map(str)
ym=data1['yearmonth']
PSOECD=data1['papr_oecd']
PSUS=data1['papr_us']
PSCA=data1['papr_ca']
PSOPECCO=data1['copr_opec']
PSFSU=data1['papr_fsu']
PSCH=data1['papr_ch']
PSW=data1['papr_world']
PCOECD=data1['patc_oecd']
PCUS=data1['patc_us']
PCCA=data1['patc_ca']
PCOECDEU=data1['patc_oecd_europe']
PCFSU=data1['patc_fsu']
PCEU=data1['patc_nonoecd_europe']
PCCH=data1['patc_ch']
PCW=data1['patc_world']
IWUS=data1['t3_stchange_us']
IWOECD=data1['t3_stchange_ooecd']
IWW=data1['t3_stchange_world']
EPIUS=data1['pasc_us']
EPIOECD=data1['pasc_oecd_t3']

ippci=pd.DataFrame({'ym':ym, 'PSOECD':PSOECD, 'PSUS':PSUS, 'PSCA':PSCA, 'PSOPECCO':PSOPECCO, 'PSFSU':PSFSU, 'PSCH':PSCH, 'PSW':PSW,
                    'PCOECD': PCOECD,'PCUS':PCUS,'PCCA':PCCA, 'PCOECDEU':PCOECDEU,'PCFSU':PCFSU, 'PCEU':PCEU, 'PCCH':PCCH, 'PCW':PCW,
                    'IWUS':IWUS,'IWOECD':IWOECD,'IWW':IWW,'EPIUS':EPIUS,'EPIOECD':EPIOECD})
ippci
```

```
data2=pd.read_excel(url1, sheet_name='3btab', header=[2,3], index_col=0).T.reset_index().drop(0)
data2.level_1 = data2.level_1.map(m)
data2['yearmonth']=data2['level_0'].map(str)+'m'+data2['level_1'].map(str)
ym=data2['yearmonth']
PSNA=data2['t3b_papr_r01']
PSRU=data2['papr_RS']
PSME=data2['t3b_papr_r05']
psnopec=pd.DataFrame({'ym':ym, 'PSNA':PSNA, 'PSRU':PSRU, 'PSME':PSME})
psnopec
```

```
data3=pd.read_excel(url1, sheet_name='3ctab', header=[2,3], index_col=0).T.reset_index().drop(0)
data3.level_1 = data3.level_1.map(m)
data3['yearmonth']=data3['level_0'].map(str)+'m'+data3['level_1'].map(str)
ym=data3['yearmonth']
PCOPEC=data3['copc_opec']
SPCOPEC=data3['cops_opec']

psopec=pd.DataFrame({'ym':ym, 'PCOPEC':PCOPEC, 'SPCOPEC':SPCOPEC})
psopec
```

```
data4=pd.read_excel(url1, sheet_name='3dtab', header=[2,3], index_col=0).T.reset_index().drop(0)
data4.level_1 = data4.level_1.map(m)
data4['yearmonth']=data4['level_0'].map(str)+'m'+data4['level_1'].map(str)
ym=data4['yearmonth']
PCNA=data4['patc_r01']
PCER=data4['patc_r03']
PCRU=data4['patc_rs']
PCME=data4['patc_r05']
WIPC=data4['rgdpq_world_pct']
OIPC=data4['rgdpq_oecd_pct']
NOIPC=data4['rgdpq_nonoecd_pct']
USDEPC=data4['forex_world_pct']

pcw=pd.DataFrame({'ym':ym, 'PCNA':PCNA, 'PCER':PCER,'PCRU':PCRU,'PCME':PCME, 'WIPC':WIPC,'OIPC':OIPC,'NOIPC':NOIPC,'USDEPC':USDEPC })
pcw
```

```
url2='https://www.eia.gov/outlooks/steo/realprices/real_prices.xlsx'
date1=['Month']
data5=pd.read_excel(url2, sheet_name='Crude Oil-M', header=39, usecols=[0,1,2,3] , parse_dates=date1,infer_datetime_format=True)
monthyear='2014-01-01'
data5=data5.dropna()
data5['Month']=pd.to_datetime(data5['Month'])
data5=data5.loc[data5['Month']>=monthyear]
data5['year'] = data5['Month'].dt.year
data5['month'] = data5['Month'].dt.month
data5['yearmonth']=data5['year'].map(str)+'m'+data5['month'].map(str)
yearmonth=data5['yearmonth']
CPI=data5['Index (1982-84=1)']
NCOP=data5['Nominal']
COP=data5['Real']
op2=pd.DataFrame({'yearmonth':yearmonth, 'CPI':CPI, 'NCOP':NCOP, 'COP':COP})

op2
```

```
url2='https://www.eia.gov/outlooks/steo/realprices/real_prices.xlsx'
date1=['Month']
data5=pd.read_excel(url2, sheet_name='Crude Oil-M', header=39, usecols=[0,1,2,3] , parse_dates=date1,infer_datetime_format=True)
monthyear='2014-01-01'
data5=data5.dropna()
data5['Month']=pd.to_datetime(data5['Month'])
data5=data5.loc[data5['Month']>=monthyear]
data5['year'] = data5['Month'].dt.year
data5['month'] = data5['Month'].dt.month
data5['yearmonth']=data5['year'].map(str)+'m'+data5['month'].map(str)
yearmonth=data5['yearmonth']
CPI=data5['Index (1982-84=1)']
NCOP=data5['Nominal']
COP=data5['Real']
op2=pd.DataFrame({'yearmonth':yearmonth, 'CPI':CPI, 'NCOP':NCOP, 'COP':COP}).groupby('yearmonth').mean()

op2
```
